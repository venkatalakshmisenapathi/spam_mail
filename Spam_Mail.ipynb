Importing the Libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.feature_extraction.text import TfidfVectorizer
```

DataCollection and Preprocessing

```python
dataset = pd.read_csv('/content/mail_data.csv')
```

```python
dataset.head()
```

```python
dataset.tail()
```

```python
dataset['Category'].value_counts()
```

```python
#replacing null values with string
dataset = dataset.where((pd.notnull(dataset)),'')
```

Label Encoding



*   Spam --> 0
*   Ham --> 1



```python
dataset.loc[dataset['Category']=='spam','Category'] = 0
dataset.loc[dataset['Category']=='ham','Category'] = 1
```

```python
dataset['Category'].value_counts()
```

```python
X = dataset['Message']
Y = dataset['Category']
```

Spliting the dataset

Feature Extraction

```python
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size = 0.2,random_state=2)
```

```python
X_train.shape,y_train.shape
```

```python
X_test.shape,y_test.shape
```

```python
feature = TfidfVectorizer(min_df=1, stop_words = 'english',lowercase=True)
```

```python
X_train_feature = feature.fit_transform(X_train)
X_test_feature = feature.transform(X_test)
```

```python
y_train = y_train.astype('int')
y_test = y_test.astype('int')
```

```python
print(X_train_feature)
```

Training Model

```python
model = LogisticRegression()
```

```python
model.fit(X_train_feature,y_train)
```

Model Prediction

```python
X_train_pred = model.predict(X_train_feature)
train_accuracy = accuracy_score(X_train_pred,y_train)
print('Accuracy Score on training data is',round(train_accuracy*100,2))
```

```python
X_test_pred = model.predict(X_test_feature)
test_accuracy = accuracy_score(X_test_pred,y_test)
print('Accuracy Score on testing data is',round(test_accuracy*100,2))
```

predictive System

```python
user_input=input("Enter the content of the mail:")
user_feature = feature.transform([user_input])
prediction = model.predict(user_feature)
if prediction[0] == 1:
  print("Ham")
else:
  print("Spam")
```
